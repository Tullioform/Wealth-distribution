## **I.a La meccanica statistica** <a id=t2> <frame src="t2" target="_self" name=t2>
La *meccanica statistica* è la branca della fisica che utilizzando la teoria della probabilità studia il comportamento meccanico e/o termodinamico di un sistema contente un gran numero di particelle. Da un punto di vista classico, lo studio di un sistema di N particelle **non** interagenti tra di loro richiede la risoluzione di N equazioni differenziali che descrivono il moto di ogni particella. Se il numero N è dell'ordine della costante di Avogadro, la meccanica statistica permette la caratterizzazione del sistema attraverso grandezze come temperatura, energia libera, pressione o volume.
Il <font color="navy">*postulato dell'equiprobabilità a priori* <font color="black"> è alla base della trattazione della materia e afferma che: <font color="navy"> *dato un sistema isolato in equilibrio, ogni **microstato** ha eguale probabilità di manifestarsi* . \
<font color="black"> Dati  Ω microstati associati ad una particolare energia, allora, la probabilità di trovare il sistema in un particolare microstato è dunque:
    $P={1\over Ω}$\
Questo postulato è necessario, perché permette di concludere che ,per un sistema in equilibrio, lo stato termdinamico che può generarsi dal maggior numero di microstati è anche lo stato più probabile del sistema.  \
La costante fondamentale della meccanica statistica è la **costante di Boltzmann** $k_B={ 1,40 × 10^{-23} JK^{-1}}$ che stabilisce la corrispondenza tra grandezze della meccanica statistica e quelle della termodinamica, come ad esempio nel caso della probabilità di uno stato e l'entropia: \
    $S={k_B log(W)}$

## **II.a Modelli Omogenei**
Partiamo da un semplice esempio: assumiamo di avere N unità interagenti {i}: i=1,2,...,N (possono essere, ad esempio, molecole di un gas) e che non vi sia energia di interazione. Denotiamo {$w_i$} le energie cinetiche con la condizione $w_i \geq \ 0$ . Una serie di aggiornamenti sull'energia cinetica $w_i(t)$ viene eseguita per tempi discreti t=0,1... e ogni aggiornamento deve tenere in conto degli effetti delle collisioni tra due molecole,che possiamo porre con uno step di ogni Δt=1, senza perdita di generalità. Tale Δt rappresenta il tempo medio che intercorre tra lo scattering di due molecole i e j. L'evoluzione del sistema può essere tracciata step by step all'istante t al seguente modo:

1. Si sceglie randomicamente una coppia di molecole distinte i e j ($1\leq\ i$ , $j\leq\ N$) con energie cinetiche $w_i$ e $w_j$ rispettivamente.
2. Computiamo il valore Δ$w_i$$j$ di energia cinetica che viene scambiata tra le molecole; questo valore sarà una funzione delle energie cinetiche iniziali $w_i$ e $w_j$ e dipende dal modello considerato.
3. Effettuiamo lo scambio energetico vero e proprio considerando:\
$w_{ij}  -> w_i - Δw_{ij}$ \
$w_j  -> w_j + Δw{ij}$ 


4. Mandare t->t+1 e tornare allo step 1

La forma della funzione Δw<sub>ij</sub> definisce il modello specifico. Vogliamo descrivere la dinamica a livello miscroscopico , basandoci sulle singole collisioni molecolari. Di fatto, descrivendo l'andamento dell'energia (o della velocità) ci siamo ridotti al solo grado di libertà w piuttosto che al set 2d delle posizioni e dei momenti delle singole particelle. Consideriamo, a titolo d'esempio la seguente regola di reshuffling:\
$w_i$ -> ε($w_i$ + $w_j$)     <font color="navy"> **[1.a]** <a id="f1"> <frame src="_self" name="f1">


$w_j$ -> (1-ε)($w_i$ + $w_j$) <font color="navy">**[1.b]**

dove ε è una variabile random interna all'intervallo [0.1].
Questa legge corrisponde a:\
$Δw_{ij}= {(1-ε)w_i -εw_j}$
Partendo da una condizione iniziale arbitraria si arriva alla distribuzione di energia all'equilibrio di Boltzmann-Gibbs,\
  $f(w)= {βe^{βw}}$ <a id="fbg"> <frame src="_self" name="fbg"> </a>
con $β ={ 1\over <w>}$ e <w> che rappresenta l'energia media di una singola molecola.
Come esempio più generale, consideriamo il rilascio di energia di un gas in d dimensioni. Assumiamo d>1 perché le distribuzioni di energia e quantità di moto di un gas mono-dimensionale non cambia nel tempo (poiché vi sono soltanto collisioni frontali). Seppure il modello possa essere sviluppato più semplicemente in 3 dimensioni, il caso di più interesse è quello dove studiamo la legge di distribuzione delle velocità in un numero arbitrario di dimensioni. Come nel caso semplificato visto prima, il nostro interesse cade  sulla forma di Δw<sub>ij</sub> che può essere dedotta dalla conservazione di energia e quantità di moto in una collisione tra due particelle distinte i e j. Denominate **$p_i$** e **$p_j$** le quantità di moto agli istanti iniziali , si dimostra che:\
$Δw_ {ij}= {r_iw_i -r_jw_j}$; <font color="navy">**[2]** <a id=f2> <frame src="_self" name="f2"> </a>
  

$r_k$=$cos^2$($a_k$) (k=i,j) <font color="navy">**[3]**

dove cos($a_k$) è la proiezione della quantità di moto **$p_k$** rispetto alla quantità di moto trasferita
Δp<sub>ij</sub> = **$p_i$** -**$p_j$** , in formule:

$cos(a_k)={ (p_k , Δp_ij) \over |p_k||Δp_ij|}$ <font color="navy">**[4]**

Le direzioni delle due particelle incidenti possono essere assunte casuali usando l'ipotesi del caos molecolare.
Fatto ciò, possiamo studiare l'evoluzione temporale del sistema, scegliendo per ogni step temporale due nuovi valori per $r_k$ per l'equazione  <font color="navy"> **[<a href=#f2>2</a>]**  <font color="black"> così da aggiornare i valori delle energie cinetiche del sistema. Usando poi le Eq. <font color="navy"> **[<a href=#f1>1.a,1.b</a>]** <font color="black">possiamo aggiornare la lista delle energie cinetiche delle particelle i e j.  Notiamo che gli $r_k$ non sono distribuiti uniformemente tra (0,1) ma il loro range è da determinarsi in base alla dimensione dello spazio d, in particolare il valore medio sarà $<r_k> = \frac{1}{d}$. 
Questa dipendenza può essere dedotta sperimentalmente: maggiore il valore di d, più improbabile sarà che $r_k$ assuma un valore vicino all'unità (cosa che invece sarà versosimile nel caso unidimensionale). Perciò. una scelta efficiente per l'intervallo dove far variare $r_k$ è (0,2/d).
Simulando questo modello , per d=2 otteniamo la funzione di Boltzmann-Gibbs: $f(w)= {βe^{-βw}}$. Per d>2 , si ottiene la generalizzazione d-dimensionale della distribuzione di Boltzmann standard, ovvero la distribuzione Γ, caratterizzata da un parametro α= d/2 :

$$f(w,α,ϑ)= {w^{α-1}e^{-w/ϑ}\over ϑ^αΓ(a)}$$ <font color="navy"> **[5]** <a id=f5> <frame src="_self" name="f5"> </a>
    
con $ϑ = \frac{<w>}{a}$ 

<font color="navy"> **[6]**
    
Dal postulato di equipartizione, risulterà
   
$$w={d k_BT\over 2}$$

Da questa relazione, guardando alla Eq <font color="navy"> **[6]** <font color="black"> possiamo identificare il parametro di scala $ϑ ={k_B T}= \frac{1}{β}$ 

Perciò il fattore di Boltzman exp(-$w$/ϑ), è presente all'interno della distribuzione di equilibrio indipendentemente dalla dimensione d , mentre il fattore $w^{α-1}$ dipende da d perché tiene conto del volume dello spazio delle fasi proporzionale a $p^{d} ∝ w^\frac{d}{2}$, dove p è il modulo della quantità di moto.

```python
#dato che le energie devono essere >= 0, creiamo una classe di particelle per stare sicuri

class Energy:
    def __init__(self, value):
        self.value = value
        self.validate(value)
        self.update(value)
        
    def validate(self, value):
        if value < 0:
            raise Exception("La particella ha energia negativa, qualcosa è andato storto")
    
    def update(self, value):
        try:
            self.validate(value)
            self.value = value
            
        except Exception as e:
            print(f"Error: {e}") 
            
class particle:
    def __init__(self, en):
        self.energy = Energy(en)

    def set_energy(self, value):
        return self.energy.update(value)
```

```python
#dato che exchange dipende da delta_energy, epsilon e choice

dependencies = {}

def dependent_on(*funcs):
    def decorator(main_func):
        dependencies[main_func.__name__] = [func.__name__ for func in funcs]
        
        @functools.wraps(main_func)
        def wrapper(*args, **kwargs):
            return main_func(*args, **kwargs)
        
        return wrapper
    return decorator

def update_dependencies(func_name):
    for main_func, deps in dependencies.items():
        if func_name in deps:
            globals()[main_func] = globals()[main_func]

def update_exchange(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        update_dependencies(func.__name__)
        return result
    return wrapper
```

```python
#definiamo ora funzioni che creino la particella inizialmente con energia random,
# e un set di queste

def mean(x):
    return np.mean(x)

def random_part(x):
    energy = np.random.rand() + mean(x)
    part = particle(energy)
    return part

def part_set(N, x):
    particles = list((random_part(x) for _ in range(N)))
    return particles
```

```python
#definiamo funzioni che scelgono particelle dal set e fanno un exchange di energia

#applico il decoratore

@update_exchange
def choice(particles):
    x = np.random.choice((len(particles)), 2)
    return x

@update_exchange
def delta_energy(particles, choices, e):
    pass

@update_exchange
def epsilon():
    pass

@dependent_on(choice, epsilon, delta_energy)
def exchange(particles):
    choices = choice(particles)
    e = epsilon()
    net_energy = delta_energy(particles, choices, e)
    new_energy_0 = particles[choices[0]].energy.value + net_energy
    new_energy_1 = particles[choices[1]].energy.value - net_energy
    
    try:
        particles[choices[0]].energy.validate(new_energy_0)
        particles[choices[1]].energy.validate(new_energy_1)
        
        particles[choices[0]].energy.value += net_energy
        particles[choices[1]].energy.value -= net_energy
        
        del new_energy_0
        del new_energy_1
    
    except:
        pass
    
    return particles
```

```python
#definiamo la funzione che fa interagire per s secondi

def interaction(particles, s):    
    t = 1
    while t < s:
        particles = exchange(particles)
        t = t + 1
    return particles
```

```python
#definisco la root mean squared tra due array

def rms(x, y):
    z = np.mean((x - y)**2)
    return np.sqrt(z)
```

## **II.a.1 Boltzmann_Gibbs**

```python
#primo esempio, non ho bisogno di ridefinire exchange per il decoratore

@update_exchange
def epsilon():
    return np.random.rand()

@update_exchange
def delta_energy(particles, choices, e):
    net_energy = (1 - e)*(particles[choices[0]].energy.value)-e*(particles[choices[1]].energy.value)
    return net_energy    
```

```python
#creo la distribuzione alla boltzmann_gibbs
def boltzmann_gibbs(x, y):
    w = np.mean(y)
    b = 1/w
    boltz = []
    for i in range(len(x)):
        boltz.append(b*np.exp(-b*x[i]))
    return boltz
```

```python
#verifichiamo che funziona
#scelgo il numero di secondi circa pi greco volte il numero di particelle, perchè sembra essere un'ottima approssimazione
N_boltz = 10000
s_boltz = 31415
#scelgo le particelle con media delle energie inizialmente nulla, per le distribuzioni successive non sarà così
particles = part_set(N_boltz, 0)

x_boltz = np.zeros(N_boltz)
prev = interaction(particles, s_boltz)

for i in range(N_boltz):
    x_boltz[i] = prev[i].energy.value
    
del prev, particles
#costruisco una distribuzione di boltzmann_gibbs ragionevole
axis_boltz = np.linspace(np.min(x_boltz), np.max(x_boltz), N_boltz)
boltz = np.array(boltzmann_gibbs(axis_boltz, x_boltz))
```

```python
#verifichiamo che sia consistente con una distribuzione alla boltzmann_gibbs
#per ora solo da punto di vista grafico

#scelgo una legge a radice per i bins, sembra andare molto bene per questo tipo di distribuzioni
bins = int(np.ceil(np.sqrt(N_boltz)))

#non indichiamo i bins sull'asse y perchè per ora non hanno significato
plt.hist(x_boltz[:], density = True, bins = bins)
plt.legend(['Simulazione'])
plt.xlabel('Energy')
plt.yticks([])
plt.show()

plt.plot(axis_boltz, boltz)
plt.xlabel('Energy')
plt.yticks([])
plt.legend(['Boltzmann_Gibbs'])

#salviamo i dati che ci serviranno dopo
boltzmann_data = [x_boltz, boltz, axis_boltz]
```

## **II.a.2. Caso n-dimensionale consistente con la distribuzione gamma**

```python
#esemipo n-dimensionale
#dobbiamo scegliere ad ogni t due nuovi valori di rk, distribuiti in maniera uniforme

@update_exchange
def epsilon():
    return 0

@update_exchange
def delta_energy(particles, choices, e):
    d = 8
    rk = np.random.uniform(0, 2/d)
    rj = np.random.uniform(0, 2/d)
    net_energy = rk*(particles[choices[0]].energy.value)-rj*(particles[choices[1]].energy.value)
    del rk, rj
    return net_energy
```

```python
#definiamo la distribuzione gamma per un numero di dimensioni d=8, lo scelgo pari in modo che la gamma mi dia un fattoriale di un intero, è solo un modo per velocizzare i calcoli
d = 8
def gamma_dist(x, d, y):
    w = np.mean(y)
    a = d/2
    theta = w/a
    y = []
    for i in range(len(x)):
        y.append((x[i]**(a-1))*np.exp(-x[i]/theta)/((theta**a)*(math.gamma(a))))
        
    return y 
```

```python
#verifichiamo che funziona
N_gamma = 10000
s = 31415
#genero le particelle con valor medio 1 perchè è ragionevole e consistente con i dati riportati sul paper
particles_gamma = part_set(N_gamma, 1)
x_gamma = np.zeros((N_gamma))
new_particles = interaction(particles_gamma, s)

for i in range(N_gamma):
    x_gamma[i] = new_particles[i].energy.value
#cancello ciò che non serve
del new_particles, particles_gamma
```

```python
#verifichiamo la consistenza con la distribuzione gamma
#dal punto di vista grafico

bins = int(np.ceil(np.sqrt(N_gamma)))

axis_gamma = np.linspace(np.min(x_gamma), np.max(x_gamma), N_gamma)
gamma = gamma_dist(axis_gamma, d, x_gamma)

plt.hist(x_gamma[:], density = True, bins = bins)
plt.legend(['Simulazione'])
plt.xlabel('Energy')
plt.yticks([])
plt.show()

plt.plot(axis_gamma, gamma)
plt.xlabel('Energy')
plt.yticks([])
plt.legend(['Gamma'])

#salviamo i dati che ci serviranno dopo
gamma_data = [x_gamma, gamma, axis_gamma]
```

## **II.b.3 Creazione delle social particles**

```python
#nel modello sociale si prevedono particelle ad "energia" negativa. Senza complicare la classe di particelle fisiche definite all'inizio, ridefiniamo una classe di "particelle sociali"


class socialenergy:
    def __init__(self, value):
        self.value = value
            
class socialparticle:
    def __init__(self, en):
        self.energy = socialenergy(en)

    def set_energy(self, value):
        return self.energy.update(value)
```

```python
#ridefiniamo anche i processi di exchange, anche se stavolta sono molto più semplici perchè non dobbiamo validare le energie

@update_exchange
def social_choice(social_particles):
    x = np.random.choice((len(social_particles)), 2)
    return x

@update_exchange
def social_lambda_energy():
    pass

@update_exchange
def social_epsilon():
    pass

@dependent_on(social_choice, social_epsilon, social_lambda_energy)
def social_exchange(social_particles):
    choices = social_choice(social_particles)
    e = social_epsilon()
    social_particles[choices[0]].energy.value = social_lambda_energy()*((social_particles[choices[0]].energy.value) + e[0]*(social_particles[choices[1]].energy.value))
    social_particles[choices[1]].energy.value = social_lambda_energy()*((social_particles[choices[1]].energy.value) + e[1]*(social_particles[choices[0]].energy.value))
    return social_particles
```

```python
#ridefiniamo le funzioni per creare social particles

def social_mean(x):
    return np.mean(x)

def random_social_part(x):
    energy = 2*np.random.rand() - 1 + social_mean(x)
    social_part = socialparticle(energy)
    return social_part

def social_part_set(N, x):
    social_particles = list((random_social_part(x) for _ in range(N)))
    return social_particles
```

```python
#definiamo la funzione che fa interagire per s secondi social particles

def social_interaction(social_particles, s):    
    t = 1
    while t < s:
        social_particles = social_exchange(social_particles)
        t = t + 1
    return social_particles
```

## **II.c Collective dynamics of opinion**
Un altro campo di applicazione della teoria cinetica dei gas è lo studio, nell'ambito dei mercati, della dinamica dell'opinione collettiva in una società. Per questo modello (ideato da **Lallalouache**), denotiamo con $w_i$ l'opinione della i-esima "particella" sociale, con $w_i$ varibale collettiva interna all'intervallo (-1,1). Il cambio nell'opinione di due agenti (che assumiamo analogo all'urto randomico di due particelle) assumerà la seguente dinamica:\
$w_i -> λ[w_i + ε w_j]$ <font color="navy"> **[11.a]** \
<font color="black"> $w_j -> λ[w_j + e w_i]$ <font color="navy"> **[11.b]** <font color="black"> \
    dove λ∈(0,1) è il parametro (assunto costante) di "convinzione" e ε,e ∈(0,1) sono parametri casuali non correlati. Notiamo che non abbiamo una conservazione graduale di opinione, a differenza di quanto accadeva nei modelli sulla ricchezza. Il dato interessante è viene a formarsi del consenso pur partendo da un set random di {$w_i$}. Si osserva una vera e propria *rottura spontanea di simmetria* nel sistema: partendo dal valor medio di $w_i$, <$w$>(0)=0 ,dato il tempo τ di rilascio del sistema , esso evolve o nello stato dove vi è una totale "opinione zero" <$w$>($t\geqτ$)=0 (per valori di $λ\leq2/3$) o nello stato di rottura di simmetria, con <$w$>($t>τ$)≠0 , dove tutti gli individui hanno opinioni o tutte posive o tutte negative ($λ\geq2/3$). Il sistema mostra una divergenza al tempo τ per $λ=λ_c={2\over3}$

### **IV.1 Metodo di Newton-Raphson per la funzione Lambert W**

```python
#possiamo stimare il punto in cui le distribuzioni esponenziali diventano di Pareto trovando la soluzione alla funzione W
#la Lambert W esce fuori dall'equazione di eulero del tipo ln(x) = ax^b

#definiamola con una tolleranza 10^-5 su 100000 iterazioni

def lambert_w_newton(z, tol=1e-5, max_iter=10000):

    w = np.log(z + 1e-20)
    for _ in range(max_iter):
        ew = np.exp(w)
        w_next = w - (w * ew - z) / (ew * (w + 1) - (w + 2) * (w * ew - z) / (2 * w + 2))
        if np.abs(w_next - w) < tol:
            return w_next
        w = w_next
    raise RuntimeError("Lambert W non converge")
```

```python
#definiamo una funzione che verifichi che la nostra distribuzione sia normalizzata secondo il modulo di matplotlib, che normalizza trattando l'istogramma come una funzione semplice e integrando

def dist_check(x, dx):
    norm = 0
    for i in range(len(x) - 1):
        norm += x[i]*np.abs(dx[i] - dx[i+1])  

    if np.abs(norm - 1) < 0.001:
        print("è una buona distribuzione")
    else:
        raise Exception("non è normalizzata!, prova ad aumentare il numero di bins")

    del norm
```

```python
#definiamo una funzione per verificare che una distribuzione sia di boltzmann. Abbiamo già i dati a disposizione
#fa già la rmse, senza dover chiamare 2 funzioni diverse

def boltzmann_check(x, y):
    b = 1/np.mean(x)
    somma = 0
    for i in range(len(x)):
        boltz = b*np.exp(-b*y[i])
        somma += ((x[i]-boltz)**2)
        somma = somma/len(x)
        boltz = 0
    return np.sqrt(somma)       
```

```python
from scipy.optimize import curve_fit
from scipy.stats import pareto, gamma
```

```python
#definiamo le funzioni gamma e pareto in un modo che ci fanno comodo
def gamma_pdf(x, a, theta):
    return (x**(a-1)*np.exp(-x/theta))/((theta**a)*(math.gamma(a)))

def gaussian_pdf(x, mu, sigma):
    return (1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*((x - mu)/(sigma))**2)

def pareto_pdf(x, a, b):
    return a*(x**(-b))
```

```python
#ora che abbiamo un range per w0, cerchiamo di stimarlo con la funzione curve_fit. per ora non definiamo la funzione che fitta, ma usiamo quelle già implementate, altrimenti non finiamo più...

w0_candidates = np.linspace(25, 26, 100)
best_params_gamma = None
best_params_pareto = None
best_w0 = None
best_log_likelihood = -np.inf

for w0 in w0_candidates:
    
    data_gamma = total_numbers[total_numbers < w0]
    data_pareto = total_numbers[total_numbers >= w0]
    
    # Verifica che abbiamo dati sufficienti in entrambi i regimi
    if len(data_gamma) < 2 or len(data_pareto) < 2:
        continue
    
    # Fit della distribuzione Gamma con parametri ragionevoli
    try:
        params_gamma, _ = curve_fit(gaussian_pdf, data_gamma, np.linspace(0, 25, 50), np.linspace(0, 3, 50))
    except RuntimeError:
        continue
    
    # Fit della distribuzione Pareto con parametri ragionevoli
    try:
        params_pareto, _ = curve_fit(pareto_pdf, data_pareto, np.linspace(0, 10, 50), np.linspace(0, 2, 50))
    except RuntimeError:
        continue
    
    # Calcola la log-verosimiglianza
    log_likelihood_gamma = np.sum(np.log(gaussian_pdf(data_gamma, *params_gamma)))
    log_likelihood_pareto = np.sum(np.log(pareto_pdf(data_pareto, *params_pareto)))
    
    log_likelihood_total = log_likelihood_gamma + log_likelihood_pareto
    
    if log_likelihood_total > best_log_likelihood:
        best_log_likelihood = log_likelihood_total
        best_w0 = w0
        best_params_gamma = params_gamma
        best_params_pareto = params_pareto

# Stampa i risultati
print("Best w0:", best_w0)
print("Gamma params:", best_params_gamma)
print("Pareto params:", best_params_pareto)
```

## **V. Altri problemi di social science**
Di seguito, esporremo altri esempi di come la teoria cinetica dei gas possa essere utilizzata per studiare i problemi già visti ma dotati di più features.\
**V.1)** Prendiamo il modello di scambio $w_i -> w_i - Δw_{ij}$  e $w_j -> w_j +Δw_{ij}$\
con $ Δw_{ij}={(1-ε)w_i - εw_j}$. Svolgendo simulazioni del modello, vogliamo trovare se viene a formarsi la distribuzione di <a href="#fbg" target="fbg"> Boltzmann-Gibbs </a> e per quale tempo si stabilizza. 

**V.2)** Different exchange rules: Andiamo a verificare che, anche considerando che per ogni step un valore arbitrario costante $Δw_i=w_0$ venga scambiato o che una frazione della ricchezza media nel sistema venga scambiata, si possa ritrovare la distribuzione di Boltzmann-Gibbs.
   
**V.3)** Tassazione delle transazioni: Abbiamo modificato le equazioni dinamiche aggiungendo "una tassa", ovvero un termine  δ$w_0$ oppure una percentuale  fissa della quantità di bene scambiato. 
    
   **V.4)** Debito: Permettiamo adesso alle particelle sociali di creare debito o fare prestiti fino ad un valore massimo $w_{max}$ così che il valore di beni possa andare in negativo ma fino ad un lower bound di -$w_{max}$. 
   
 **V.5)** Fixed Saving: Aggiungiamo al modello un valore di "saving" $w_0$ per ognuno dei due agenti, così che la ricchezza totale venga diminuita di un fattore $2w_0$ 
    
 **V.6)** Interazione tra più agenti: Studiamo il sistema dinamico della "morra cinese", al fine di usare una nuova dinamica al sistema che preveda che la ricchezza possa essere distribuita in incontri ,non solo, di due ma di 3 agenti.

### **V.4 Debt**

```python
#qui le particelle hanno un debito, ossia energia negativa. mi servono le social particles

@update_exchange
def social_choice(social_particles):
    x = np.random.choice((len(social_particles)), 2)
    return x

@update_exchange
def social_lambda_energy(social_particles, choices, e):
    wmax = 5
    #al loan eviterei di mettere come variabile casuale di nuovo epsilon
    loan = np.random.rand()*wmax
    net_energy = ((1 - e)*(social_particles[choices[0]].energy.value)-e*(social_particles[choices[1]].energy.value)) - loan
    val1 = social_particles[choices[0]].energy.value + net_energy
    val2 = social_particles[choices[1]].energy.value - net_energy
    if val1 < -wmax:
        social_particles[choices[0]].energy.value = -wmax
        social_particles[choices[1]].energy.value -= net_energy
        net_energy = 0
    else:
        pass
    if val2 < -wmax:
        social_particles[choices[1]].energy.value = -wmax
        social_particles[choices[0]].energy.value += net_energy
        net_energy = 0
    else:
        pass
    del val1, val2
    return net_energy

@update_exchange
def social_epsilon():
    return np.random.rand()

#qui mi serve proprio una nuova funzione, mi conviene ridefinirla piuttosto che modificare la precedente
@dependent_on(social_choice, social_epsilon, social_lambda_energy)
def debt_exchange(social_particles):
    choices = social_choice(social_particles)
    e = social_epsilon()
    net_social_energy = social_lambda_energy(social_particles, choices, e)
    social_particles[choices[0]].energy.value += net_social_energy
    social_particles[choices[1]].energy.value -= net_social_energy
    return social_particles
```

```python
#... e devo ridefinire l'interaction

def debt_interaction(social_particles, s):    
    t = 1
    while t < s:
        social_particles = debt_exchange(social_particles)
        t = t + 1
    return social_particles
```

```python
#immagino che dobbiamo tornare alle particles con energia > 0 
#il termine additivo mi deve per forza far ridefinire l'exchange

#in realtà i decorator non sono proprio necessari in questo punto
@update_exchange
def epsilon():
    return np.random.rand()

@update_exchange
def delta_energy(particles, choices, e):
    net_energy = (1 - e)*(particles[choices[0]].energy.value)-e*(particles[choices[1]].energy.value)
    return net_energy

#imponiamo w0 = 0.1
def saving_exchange(particles):
    choices = choice(particles)
    e = epsilon()
    w0 = 0.1
    net_energy = delta_energy(particles, choices, e)
    new_energy_0 = particles[choices[0]].energy.value + net_energy + w0
    new_energy_1 = particles[choices[1]].energy.value - net_energy + w0
    
    try:
        particles[choices[0]].energy.validate(new_energy_0)
        particles[choices[1]].energy.validate(new_energy_1)
        
        particles[choices[0]].energy.value += net_energy
        particles[choices[1]].energy.value -= net_energy
        
        particles[choices[0]].energy.value += w0
        particles[choices[1]].energy.value += w0
        
        del new_energy_0
        del new_energy_1
        #total wealth diminished  by 2w0
        for i in range(len(particles)):
            if i == choices[0] or i == choices[1]:
                pass
            else:
                particles[i].energy.value -= (2*w0)/(len(particles) - 2)
    except:
        pass
    
    return particles
```

```python
#... e devo ridefinire l'interaction

def saving_interaction(particles, s):    
    t = 1
    while t < s:
        particles = saving_exchange(particles)
        t = t + 1
    return particles
```

```python
saving_N = 10000
saving_s = 10000

#ho messo una bias a 5 perchè si evidenziano comportamenti interessanti. effettivamente la tassa di 0.1 era troppo grossa...
particles = part_set(saving_N, 5)

saving_x = np.zeros(saving_N)
prev = saving_interaction(particles, saving_s)

for i in range(saving_N):
    saving_x[i] = prev[i].energy.value
    
del prev, particles
```

```python
bins = int(np.ceil(np.sqrt(saving_N)))

plt.hist(saving_x[:], density = True, bins = bins)
plt.legend(['Simulazione'])
plt.xlabel('Energy')
plt.title('Fixed Saving')
plt.yticks([])
plt.show()
```

### **V.6 Many agent interaction: la morra cinese**
Prima di vedere in atto il modello della morra cinese nell'ambito dell'interazione di 3 corpi, cerchiamo dapprima di capirne i concetti fondamentali.
Consideriamo un sistema dinamico in cui tre popolazioni crescono proprorzionalmente al prodoto della loro numerosità moltiplicato per la seconda popolazione (che immaginiamo siano le loro prede) ma decrescono proporzionalmente alla loro numerosità moltiplicata quella della terza popolazione (che immaginiamo essere loro predatori). 
Matematicamente, il sistema dinamico è descritto da una permutazione ciclica delle variabili:\
$x'={x(y-z)}$ \
$y'={y(z-x)}$ \
$z'={z(x-y)}$ \
Per capire l'evoluzione del sistema, notiamo che $x'+y'+z'={0}$ \
ovvero $x+y+z={cost.}$\
ma anche \
$${d(xyz)\over dt} = {0}$$ \
cioè, anche il prodotto xyz è costante.
In particolare, deduciamo che se i tre valori sono tutti positivi, il moro resterà nel quadrante positivo e sarà limitato dal primo vincolo. Diventa dunque impossibile che una popolazione si estingua, poiché se un valore andasse a 0, gli altri due dovrebbero crescere p


```python
#dobbiamo definire un'interazione tra 3 particelle, imponendo che localmente la somma si conservi. Ci rifacciamo al sistema dinamico della morra cinese

#ora ne prendo 3
@update_exchange
def three_choice(particles):
    x = np.random.choice((len(particles)), 3)
    return x

#returna tutti 1 in modo da conservare tutto
@update_exchange
def three_epsilon():
    return np.ones(3)

#per ora metto a 0
@update_exchange
def three_delta_energy():
    return np.zeros(3)

@dependent_on(three_epsilon, three_choice, three_delta_energy)
def three_exchange(particles):
    choices = three_choice(particles)
    e = three_epsilon()
    net_energy = three_delta_energy()
    new_energy_0 = e[0]*particles[choices[0]].energy.value*(net_energy[0] + (particles[choices[1]].energy.value - particles[choices[2]].energy.value))
    new_energy_1 = e[1]*particles[choices[1]].energy.value*(net_energy[1] + (particles[choices[2]].energy.value - particles[choices[0]].energy.value))
    new_energy_2 = e[2]*particles[choices[2]].energy.value*(net_energy[2] + (particles[choices[0]].energy.value - particles[choices[1]].energy.value))
    try:
        particles[choices[0]].energy.validate(particles[choices[0]].energy.value + new_energy_0)
        particles[choices[1]].energy.validate(particles[choices[1]].energy.value + new_energy_1)
        particles[choices[1]].energy.validate(particles[choices[2]].energy.value + new_energy_2)
        
        particles[choices[0]].energy.value += new_energy_0
        particles[choices[1]].energy.value += new_energy_1
        particles[choices[2]].energy.value += new_energy_2
        
        del new_energy_0
        del new_energy_1
        del new_energy_2
    
    except:
        pass
    
    return particles
```

```python
#definiamo la funzione che fa interagire 3 particelle per s secondi

def three_interaction(particles, s):    
    t = 1
    while t < s:
        particles = three_exchange(particles)
        t = t + 1
    return particles
```

```python
#verifichiamo che funziona

N = 10000
s = 31415

three_particles = part_set(N, 1)

x_three = np.zeros(N)
prev = three_interaction(three_particles, s)

for i in range(N):
    x_three[i] = prev[i].energy.value
    
del prev, three_particles
```

```python
#scelgo una legge a radice per i bins, sembra andare molto bene per questo tipo di distribuzioni
bins = int(np.ceil(np.sqrt(N)))

plt.hist(x_three[:], density = True, bins = bins)
plt.legend(['Many agents interaction'])
plt.xlabel('Energy')
plt.yticks([])
plt.show()
```
